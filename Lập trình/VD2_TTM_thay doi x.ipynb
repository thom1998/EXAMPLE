```python
def proj_C(a,b,x):
    if (np. inner (a,x) <= b):
        return x
    else :
        return x + (b - np.inner(a,x))/(np.linalg.norm(a)**2)*a

def proj_Q(c,d,y):
    if (np. inner (c,y) <= d):
        return y
    else :
        return y + (d - np.inner(c,y))/(np.linalg.norm(c)**2)*c
```

```python
def A(x):
  return x/2
def B(y):
  return y - proj_Q(c, d, y)
```

```python
def y_k(x):
  return beta_k*x + (1 - beta_k)*proj_C(a, b, x - _lambda*A(x))

def z_k(y):
  return proj_Q(c, d, np.matmul(F, y_k(y)) - _lambda*B(np.matmul(F, y_k(y))))
```

```python
def v_k(x):
  gamma = ro_k*(np.linalg.norm(z_k(x) - np.matmul(F,y_k(x)))**2)/(np.linalg.norm(np.matmul(F.T,z_k(x) - np.matmul(F,y_k(x))))**2 + k_k)
  return y_k(x) + gamma*np.matmul(F.T,z_k(x) - np.matmul(F,y_k(x)))
```
