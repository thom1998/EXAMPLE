```python
def proj_C(a,b,x):
    if (np. inner (a,x) == b):
        return x
    else :
        return x + (b - np.inner(a,x))/(np.linalg.norm(a)**2)*a

def proj_Q(c,d,y):
    if (np. inner (c,y) == d):
        return y
    else :
        return y + (d - np.inner(c,y))/(np.linalg.norm(c)**2)*c
```
