```python
def proj_C(a,b,x):
    if (np. inner (a,x) <= b):
        return x
    else :
        return x + (b - np.inner(a,x))/(np.linalg.norm(a)**2)*a

def proj_Q(c,d,y):
    if (np. inner (c,y) <= d):
        return y
    else :
        return y + (d - np.inner(c,y))/(np.linalg.norm(c)**2)*c
```

```python
def F_1(x):
  return x - proj_C(a,b,x)
def F_2(x):
  return x - proj_Q(c,d,x)
```

```python
def v_k(x):
  return proj_Q(c,d,u_k(x) - muy_k*F_2(u_k(x)))
```

```python
def proj_Q_k(m,n,y):
    if (np. inner (m,y) <= n):
        return y
    else :
        return y + (n - np.inner(m,y))/(np.linalg.norm(m)**2)*m
```
